```python
labels = ["Affordable", "Mid-range", "Expensive", "Luxury"]
df["price_category"]=pd.qcut(df["price"],4,labels=labels)
```

```python
labels=["New","Like_New","Used"]
df["condition"]=pd.qcut(df["total_km_travled"],3,labels=labels)
```
